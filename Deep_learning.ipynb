```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from PIL import Image
```

```python
nomes_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck' ]
```

```python
(imagens_treino, labels_treino), (imagens_teste, labels_teste) = datasets.cifar10.load_data()
```

```python
imagens_treino = imagens_treino / 255.0
imagens_teste = imagens_teste / 255.0
```

```python
def visualização_de_imagens(imagens, labels):
  plt.figure(figsize=(10, 10))
  for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(imagens[i], cmap="gray")
    plt.xlabel(labels[i])
  plt.show()
```

visualização_de_imagens(imagens_treino, labels_teste)

```python
visualização_de_imagens (imagens_treino, labels_treino)
```

```python
modelo_dsa = tf.keras.models.Sequential()
modelo_dsa.add(tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (32, 32, 3)))
modelo_dsa.add (tf.keras.layers.MaxPooling2D((2, 2)))

modelo_dsa.add(tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add (tf.keras.layers.MaxPooling2D((2, 2)))

modelo_dsa.add(tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add (tf.keras.layers.MaxPooling2D((2, 2)))
```

```python
modelo_dsa = models.Sequential()
modelo_dsa.add(layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (32, 32, 3)))
modelo_dsa.add (layers.MaxPooling2D((2, 2)))

modelo_dsa.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add (layers.MaxPooling2D((2, 2)))

modelo_dsa.add(layers.Conv2D(64, (3, 3), activation = 'relu'))
modelo_dsa.add (layers.MaxPooling2D((2, 2)))
```

```python
modelo_dsa.add(layers.Flatten())
modelo_dsa.add(layers.Dense(64, activation = 'relu'))
modelo_dsa.add(layers.Dense(10, activation = 'softmax'))
```

```python
modelo_dsa.summary()
```

```python
modelo_dsa.compile(optimizer = 'adam',
                   loss = 'sparse_categorical_crossentropy',
                   metrics = ['accuracy'])
```

```python
# time
history = modelo_dsa.fit(imagens_treino, labels_treino, epochs = 10, validation_data = (imagens_teste, labels_teste))
```

```python
erro_teste, acc_teste = modelo_dsa.evaluate(imagens_teste, labels_teste)
```

```python
print("A acuracia do modelo foi de: ", round(acc_teste*100, 1), "%")
```
